```python
# Multiple Linear Regression
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import warnings
warnings.filterwarnings("ignore")
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn import metrics
def model(regressor,X_Train,Y_Train,X_Test,Y_Test):
#     print(regressor)
#     regressor = MultiOutputClassifier(classifier)
    scores = cross_val_score(regressor, X_Train, Y_Train, cv=5, scoring='f1_macro')
    model1=regressor.fit(X_Train,Y_Train)
#     print('training accuracy')
#     print(model1.score(X_Train, Y_Train))
    Y_pred = model1.predict(X_Test)
    from sklearn.metrics import accuracy_score
#     print('testing accuracy')
    acc1 = (accuracy_score(Y_Test, Y_pred))
#     print(model1.score(X_Test,Y_Test))
#     from sklearn.metrics import classification_report
    # print(classification_report(Y_Test, Y_pred))
    # print(Y_pred)
    return model1,Y_pred,acc1,scores
def logic(X_Train,Y_Train,X_Test,Y_Test):
    from sklearn.linear_model import LogisticRegression
    regressor = LogisticRegression()
    model1,Y_pred,acc1,scores=model(regressor,X_Train,Y_Train,X_Test,Y_Test)
    # print(pd.DataFrame(zip(X.columns, np.transpose(model1.coef_))))    
    return Y_pred,acc1,scores    
def decision(X_Train,Y_Train,X_Test,Y_Test):
    from sklearn.tree import DecisionTreeClassifier
    regressor = DecisionTreeClassifier(criterion = 'entropy', random_state = 42)
    model1,Y_pred,acc1,scores=model(regressor,X_Train,Y_Train,X_Test,Y_Test)
    # print(pd.DataFrame(zip(X.columns, np.transpose(model1.coef_))))    
    return Y_pred,acc1,scores    
def knn(X_Train,Y_Train,X_Test,Y_Test):
    from sklearn.neighbors import KNeighborsClassifier
    regressor = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
    model1,Y_pred,acc1,scores=model(regressor,X_Train,Y_Train,X_Test,Y_Test)
    # print(pd.DataFrame(zip(X.columns, np.transpose(model1.coef_))))    
    return Y_pred,acc1,scores    
def svm_kernel(X_Train,Y_Train,X_Test,Y_Test):
    from sklearn.svm import SVC
    regressor = SVC(kernel = 'rbf', random_state = 0)
    model1,Y_pred,acc1,scores=model(regressor,X_Train,Y_Train,X_Test,Y_Test)
    # print(pd.DataFrame(zip(X.columns, np.transpose(model1.coef_))))    
    return Y_pred,acc1,scores    
def naive(X_Train,Y_Train,X_Test,Y_Test):
    from sklearn.naive_bayes import GaussianNB
    regressor = GaussianNB()
    model1,Y_pred,acc1,scores=model(regressor,X_Train,Y_Train,X_Test,Y_Test)
    # print(pd.DataFrame(zip(X.columns, np.transpose(model1.coef_))))    
    return Y_pred,acc1,scores    
def svm(X_Train,Y_Train,X_Test,Y_Test):
    from sklearn.svm import SVC
    regressor = SVC(kernel = 'linear', random_state = 0)
    model1,Y_pred,acc1,scores=model(regressor,X_Train,Y_Train,X_Test,Y_Test)
    # print(pd.DataFrame(zip(X.columns, np.transpose(model1.coef_))))    
    return Y_pred,acc1,scores    
def random(X_Train,Y_Train,X_Test,Y_Test):
    from sklearn.ensemble import RandomForestClassifier
    regressor = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
    model1,Y_pred,acc1,scores=model(regressor,X_Train,Y_Train,X_Test,Y_Test)
    # print(pd.DataFrame(zip(X.columns, np.transpose(model1.coef_))))    
    return Y_pred,acc1,scores    
```
